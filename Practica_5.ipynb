```python
import pandas as pd

# URL del archivo CSV
data_url = "https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv"

# Cargar datos en un DataFrame
df = pd.read_csv(data_url)

def buscar_pelicula(nombre_pelicula):
    resultado = df[df['title'].str.lower() == nombre_pelicula.lower()]
    if resultado.empty:
        print("Película no encontrada.")
    else:
        print(resultado)

# Solicitar nombre de la película al usuario
nombre_pelicula = input("Ingrese el nombre de la película: ")
buscar_pelicula(nombre_pelicula)
```

```python
import pandas as pd

# URL del archivo CSV
data_url = "https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv"

# Cargar datos en un DataFrame
df = pd.read_csv(data_url)

# Mostrar la cantidad de filas y columnas
print(f"El DataFrame tiene {df.shape[0]} filas y {df.shape[1]} columnas.")

# Mostrar los nombres de las columnas y sus tipos de datos
print("\nInformación de las columnas:")
print(df.dtypes)

def buscar_pelicula(nombre_pelicula):
    resultado = df[df['title'].str.lower() == nombre_pelicula.lower()]
    if resultado.empty:
        print("Película no encontrada.")
    else:
        print(resultado)

# Solicitar nombre de la película al usuario
nombre_pelicula = input("Ingrese el nombre de la película: ")
buscar_pelicula(nombre_pelicula)
```
